```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
import sys

PROJECT_ROOT = Path().resolve().parent
sys.path.append(str(PROJECT_ROOT / "src"))

from data_ingest import load_raw_email_data

sns.set(style="whitegrid", palette="muted")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
df = load_raw_email_data()
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all').T
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
num_cols
```

```python
df[num_cols].hist(bins=30, figsize=(14,12))
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
cat_cols
```

```python
for col in cat_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().head(20).plot(kind='bar')
    plt.title(col)
    plt.show()
```

```python
funnel_counts = {
    "Sent": len(df),
    "Opened": df['open_flag'].sum(),
    "Clicked": df['click_flag'].sum(),
    "Converted": df['conversion_flag'].sum()
}
funnel_counts
```

```python
plt.figure(figsize=(8,5))
plt.bar(funnel_counts.keys(), funnel_counts.values(), color=['grey', 'blue', 'green', 'purple'])
plt.title("Email Marketing Funnel")
plt.ylabel("Users")
plt.show()
```

```python
df['consumer_archetypes'].value_counts().plot(kind="bar", figsize=(8,4))
plt.title("Distribution of Consumer Archetypes")
plt.show()
```

```python
archetype_open = df.groupby("consumer_archetypes")['open_flag'].mean().sort_values()
archetype_open.plot(kind="barh")
plt.title("Open Rate by Consumer Archetypes")
plt.xlabel("Open Rate")
plt.show()
```

```python
archetype_click = df.groupby("consumer_archetypes")['click_flag'].mean().sort_values()
archetype_click.plot(kind="barh")
plt.title("Click Rate by Consumer Archetype")
plt.xlabel("Click Rate")
plt.show()
```

```python
df.groupby("consumer_archetypes")['conversion_flag'].mean().sort_values().plot(kind="barh")
plt.title("Conversion Rate by Archetype")
plt.show()
```

```python
df["mosaic_segment"].value_counts().plot(kind="bar", figsize=(8,4))
plt.title("Mosaic Segment Distribution")
plt.show()
```

```python
df.groupby("mosaic_segment")["open_flag"].mean().plot(kind="bar")
plt.title("Open Rate by Mosaic Segment")
plt.show()
```

```python
df["mailing_category"].value_counts().plot(kind="bar", figsize=(6,4))
plt.title("Mailing Category Distribution")
plt.show()
```

```python
df.groupby("mailing_category")["open_flag"].mean().plot(kind="barh")
plt.title("Open Rate by Campaign Category")
plt.show()
```

```python
df.groupby("mailing_category")['click_flag'].mean().plot(kind="bar")
plt.title("Click Rate by Campaign Category")
plt.show()
```

```python
df.groupby("device_type")["click_flag"].mean().plot(kind="barh")
plt.title("Click Rate by Device Type")
plt.show()
```

```python
df.groupby("mailing_hour")["open_flag"].mean().plot()
plt.title("Open Rate by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Open Rate")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Key Insights:")
print("-------------")
print(f"Overall open rate : {df['open_flag'].mean():.3f}")
print(f"Overall click rate : {df['click_flag'].mean():.3f}")
print(f"Overall conversion rate : {df['conversion_flag'].mean():.3f}")
print(f"Highest Opening Archetype : {df.groupby('consumer_archetypes')['open_flag'].mean().idxmax()}")
print(f"Highest Converting Archetype : {df.groupby('consumer_archetypes')['conversion_flag'].mean().idxmax()}")
```
